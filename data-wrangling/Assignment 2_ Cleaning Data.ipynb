###**Project 6-1: Clean the Ramen Data**

```python
# Load ramen data
ramen = pd.read_csv("ramen-ratings.csv")
print("Original rows:", len(ramen))
ramen.head()
```

```python
# 1. Rename Stars → Rating
ramen.rename(columns={"Stars": "Rating"}, inplace=True)
```

```python
# 2. Convert Style column to category type
ramen["Style"] = ramen["Style"].astype("category")
```

```python
# 3. Drop the Country column
ramen.drop(columns=["Country"], inplace=True)
```

```python
# 4. Check changes
print(ramen.info())
```

```python
# 5. Rename Brand → Company, Variety → Product
ramen.rename(columns={"Brand": "Company", "Variety": "Product"}, inplace=True)
```

```python
# 6. Show duplicates on Company/Product
dupes = ramen[ramen.duplicated(subset=["Company","Product"], keep=False)]
print("Duplicate Company/Product rows:\n", dupes)
```

```python
# 7. Drop duplicates except first
ramen_nodupes = ramen.drop_duplicates(subset=["Company","Product"], keep="first")
print("Rows after removing duplicates:", len(ramen_nodupes))
```

```python
# 8. Drop missing values
ramen_clean = ramen_nodupes.dropna()
print("Final rows after dropping NA:", len(ramen_clean))
ramen_clean.head()
```

###**Project 6-2: Clean the Avocado Data**

```python
# Load avocado data
avocado = pd.read_csv("avocado.csv")
avocado.head()
```

```python
# 1. Check missing values
print(avocado.isnull().sum())
```

```python
# 2. Unique values in region column
print("Unique regions:", avocado["region"].unique())
```

```python
# 3. Drop rows where region == "TotalUS"
avocado = avocado[avocado["region"] != "TotalUS"]
print("Rows after dropping TotalUS:", len(avocado))
```

```python
# 4. Drop unwanted columns
avocado = avocado.drop(columns=["Unnamed: 0","4046","4225","4770"], errors="ignore")
```

```python
# 5. Convert Date column to datetime
avocado["Date"] = pd.to_datetime(avocado["Date"])
```

```python
# 6. Check final DataFrame
print(avocado.info())
avocado.head()
```
